## Reading PM2.5 data from TileDB on Azure

```python
%matplotlib inline
import tiledb
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

To access TileDB data on Azure, we need to provide Azure account details to the tiledb "ctx" object.

In the below cell, fill in the correct value for your storage account key - this can be found on the Azure portal by finding the "storage account" (called "dimaqdata" in our case) and clicking on "Access Keys" in the left sidebar.   You then need to click "show keys" at the top of the page to un-obfuscate the keys.


```python
config = tiledb.Config()
config["vfs.azure.storage_account_name"] = "dimaqdata"
config["vfs.azure.storage_account_key"] = <INSERT_AZURE_ACCESS_KEY_HERE>
config["vfs.azure.blob_endpoint"] = "dimaqdata.blob.core.windows.net"
ctx = tiledb.Ctx(config=config)
```

The name of the "container" with our tiledb data is "who2021-tiledb".  Within this, there are currently datasets called "PMSummariesGrid" and "PMSummariesCountry".

Look at the grid first:

```python
tiledb_url = "azure://who2021-tiledb/PMSummariesGrid"
A = tiledb.open(tiledb_url, ctx=ctx)
```

To test that this worked, and that we can slice the tiledb array in our three dimensions ("Time", "Longitude", "Latitude") let's make a quick plot of Mean PM2.5 in 2014 in the lat/long region of the UK:

```python
gb = A.df[2014, -5.:5.,50.:60.]
p = gb.plot.scatter("Longitude", "Latitude", c="Mean")
```

Now we can look at the "Country" level summary:

```python
tiledb_url = "azure://who2021-tiledb/PMSummariesCountry"
A = tiledb.open(tiledb_url, ctx=ctx)
```

Here we have the dimensions "CountryCode", "Year", and "Type", where "Type" refers to whether the numbers are weighted or unweighted.

Let's get the population weighted data for all years, for the country with country-code 44:

```python
df = A.df[44, :,"Population-weighted"]
```

```python
df
```

```python
df.plot(x="Year",y="Mean")
```

Now try the median for the first 20 country codes, for unweighted data:

```python
df = A.df[:20, :,"Unweighted"]
```

Use the seaborn package as an easy way of plotting lots of lines together:

```python
sns.lineplot(data=df, x='Year', y='Median', hue='CountryCode')
```
